```python
import requests
from bs4 import BeautifulSoup
import csv

# Define the URL of the BigBasket page you want to scrape
url = "https://www.bigbasket.com/pc/fruits-vegetables/fresh-vegetables/"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page
soup = BeautifulSoup(response.content, "html.parser")

# Find the elements containing the product information
products = soup.find_all("div", class_="product-item")

# Create lists to store the extracted data
product_names = []
product_prices = []
product_links = []

# Extract data from each product
for product in products:
    # Extract product name
    product_name = product.find("a", class_="product-name").text.strip()
    product_names.append(product_name)

    # Extract product price
    product_price = product.find("span", class_="disc-price").text.strip()
    product_prices.append(product_price)

    # Extract product link
    product_link = product.find("a", class_="product-name")["href"]
    product_links.append("https://www.bigbasket.com" + product_link)

# Combine the data into a list of tuples
product_data = list(zip(product_names, product_prices, product_links))

# Save the data to a CSV file
with open("bigbasket_vegetable_data.csv", "w", newline="") as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(["Product Name", "Product Price", "Product Link"])  # Write a header row
    writer.writerows(product_data)

print("Data scraped from BigBasket and saved to bigbasket_vegetable_data.csv")
```

```python
import requests
from bs4 import BeautifulSoup
from tabulate import tabulate

url = "https://www.flipkart.com/food-products/vegetables/fresh-vegetables/pr?sid=eat%2Clmt%2Czhh"
r = requests.get(url)
soup = BeautifulSoup(r.content, "html.parser")
product = soup.find_all('a', class_='s1Q9rs')
price = soup.find_all('div', class_='_30jeq3')
image = soup.find_all('img', class_='_396cs4')  # Locate the img tags containing product images

# Create lists to store the extracted data
product_names = [p.text for p in product]
product_prices = [p.text.replace('â‚¹', '') for p in price]
product_links = [p['href'] for p in product]
image_links = [img['src'] for img in image]  # Extract the src attribute of the img tags

# Combine the data into a list of tuples
product_data = list(zip(product_names, product_prices, product_links, image_links))

# Display the data in a table
headers = ['Product Name', 'Product Price', 'Product Link', 'Image Link']
df = tabulate(product_data, headers=headers, tablefmt='pretty')
print(df)
```

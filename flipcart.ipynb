```python
import requests
from bs4 import BeautifulSoup
from tabulate import tabulate

url = "https://www.flipkart.com/food-products/vegetables/fresh-vegetables/pr?sid=eat%2Clmt%2Czhh"
r = requests.get(url)
soup = BeautifulSoup(r.content, "html.parser")
product = soup.find_all('a', class_='s1Q9rs')
price = soup.find_all('div', class_='_30jeq3')

# Create lists to store the extracted data
Product_names = [p.text for p in product]
Product_prices = [p.text.replace('₹', '') for p in price]
Source=['flip']

links = [p['href'] for p in product]
```

```python
product_data = list(zip(Product_names, Product_prices,Source,links))
```

```python
product_data
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(product_data)

df.columns = ['Product_name', 'Product_price','Source', 'links']

# Now, your DataFrame will have the desired column names
print(df)
```
